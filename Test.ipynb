```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = pd.DataFrame({
    'age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80],
    'income': [40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000, 100000]
})

# Histogram
plt.figure(figsize=(10, 5))
plt.hist(data['age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Boxplot
plt.figure(figsize=(10, 5))
plt.boxplot(data['income'], vert=False)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = pd.DataFrame({
    'age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80],
    'income': [40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000, 100000]
})

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['income'], color='green', alpha=0.7)
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (epochs vs loss)
epochs = range(1, 11)
train_loss = [0.5, 0.4, 0.3, 0.25, 0.2, 0.18, 0.15, 0.12, 0.1, 0.08]
val_loss = [0.6, 0.5, 0.4, 0.35, 0.3, 0.28, 0.25, 0.22, 0.2, 0.18]

# Plot learning curve
plt.figure(figsize=(8, 6))
plt.plot(epochs, train_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.title('Learning Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Example data
y_true = [0, 1, 1, 0, 1, 0, 1, 0]
y_pred = [0, 1, 0, 0, 1, 1, 1, 0]

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
y_true = [3, 5, 7, 9, 11]
y_pred = [2.8, 5.1, 7.2, 8.9, 11.5]

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, color='blue')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='red', linestyle='--')  # Diagonal line
plt.title('Predicted vs Actual')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()  
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Example data
X = [[0, 1], [1, 0], [1, 1], [0, 0]]
y = [0, 1, 1, 0]

# Train model
model = RandomForestClassifier()
model.fit(X, y)

# Feature importance
importance = model.feature_importances_
features = ['Feature 1', 'Feature 2']

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(features, importance, color='green')
plt.title('Feature Importance')
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Example data
X = [[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]]

# K-Means clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
labels = kmeans.labels_

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter([x[0] for x in X], [x[1] for x in X], c=labels, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
